# Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Import Dataset

```python
train=pd.read_csv('train.csv',low_memory=False)
test=pd.read_csv('test.csv',low_memory=False)
store=pd.read_csv('store.csv',low_memory=False)
train.head()
```

```python
train['Date'] = pd.to_datetime(train['Date'])
selected = [1, 2, 3, 4]
train_subset = train[train['Store'].isin(selected)].copy()
train_weekly = train_subset.set_index('Date').groupby('Store')['Sales'].resample('W').sum().reset_index()
plt.figure(figsize=(15, 7))
sns.set_style("white")
sns.lineplot(data=train_weekly, x='Date', y='Sales', hue='Store', palette='tab10', linewidth=1)

plt.xlabel('Date', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.grid(axis='y', linestyle='-', alpha=0.3)
plt.legend(title='Store', frameon=True, shadow=True)

plt.tight_layout()
plt.show()
```

```python
hambastegi= train.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(hambastegi, annot=True,fmt=".2f",cmap='RdYlGn',linewidths=0.5,center=0)
plt.show()
```

```python
train_store= pd.merge(train,store, on='Store')
```

```python
train_store.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Promo',y='Sales',data=train_store)
plt.xticks([0, 1], ['no promo','promo'])
plt.show()
```

```python
sns.pointplot(x='DayOfWeek',y='Sales',hue='SchoolHoliday',data=train_store)
plt.show()
```

```python
train_store['StateHoliday'] = train_store['StateHoliday'].astype(str)
open_stores=train_store[train_store['Open']==1]
holiday_impact= open_stores.groupby('StateHoliday')['Sales'].mean().reset_index()
sns.barplot(x='StateHoliday',y='Sales',data=holiday_impact)
plt.xticks([0, 1, 2, 3], ['Normal Day', 'Public Holiday', 'Easter', 'Christmas'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='StoreType', y='Sales', data=open_stores, order=['a','b','c','d'])
plt.show()
```
